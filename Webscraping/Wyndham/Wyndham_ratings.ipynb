```python
url = 'https://www.wyndhamhotels.com//microtel/albertville-alabama/microtel-inn-and-suites-of-albertville/overview'
res = requests.get(url)
soup = BeautifulSoup(res.content, 'lxml')
testurl = soup.find('iframe')['data-src']
urlres = requests.get(testurl)
urlsoup = BeautifulSoup(urlres.content, 'lxml')
rating = urlsoup.find('body', {'id' : 'WIDGET'}).find('img').attrs['src']
```

```python
rating
```

```python
dataframe = hotels
```

```python
%%time

hotelratings = []
counter = 1
for index, row, in dataframe.iterrows():
    name = row['name']
    url = row['url']
    try:
        res = requests.get(url)
        soup = BeautifulSoup(res.content, 'lxml')
        testurl = soup.find('iframe')['data-src']
        urlres = requests.get(testurl)
        urlsoup = BeautifulSoup(urlres.content, 'lxml')
        rating = urlsoup.find('body', {'id' : 'WIDGET'}).find('img').attrs['src']
        hotelrating = {}
        hotelrating['name'] = name
        hotelrating['rating'] = rating
        hotelratings.append(hotelrating)
        time = datetime.now()
        print(f'{counter} hotels at {time}')
    except:
        print(f'{name} unsuccessful')
        pass
    counter += 1
hotelratings_df = pd.DataFrame(hotelratings)
dataframe1 = pd.merge(dataframe, hotelratings_df, on='name')
dataframe1
```
